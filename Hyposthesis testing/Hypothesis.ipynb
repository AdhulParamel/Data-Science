Estimation and Confidence intervals


```python
## a
```

```python
import numpy as np
import scipy.stats as stats
n=15
data=[1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
x_bar=np.mean(data)
```

```python
x_bar
```

```python
t_critical= stats.t.ppf(0.995,14)
```

```python
t_critical
```

```python
std=np.std(data,ddof=1)
std
```

```python
Margin_of_Error= t_critical*(std/np.sqrt(n))
Margin_of_Error
```

```python
pop_mean_lower= x_bar-Margin_of_Error
pop_mean_upper= x_bar+Margin_of_Error
print(pop_mean_lower,',',pop_mean_upper)
```

```python
s=0.2
```

```python
z_crit = stats.norm.ppf(0.995)
z_crit
```

```python
margin= z_crit*(s/np.sqrt(n))
margin
```

```python
mean_lower= x_bar-margin
mean_upper= x_bar+margin
print(mean_lower,',',mean_upper)
```
