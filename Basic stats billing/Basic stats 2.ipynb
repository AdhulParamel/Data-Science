```python
df2=pd.read_csv('Billing_Data.csv')
```

```python
## 2 Select columns
a=df1[['PatientID','Department','Doctor','BillAmount']]
a
```

```python
##3 drop 
df1.drop(columns=['ReceptionistID','CheckInTime'])
```

```python
##4 Total bill amount of each department
a.groupby('Department')['BillAmount'].sum()
```

```python
a.duplicated().sum()
```

```python
##5 drop duplicates
a.drop_duplicates(inplace=True)
```

```python
a.groupby('Department')['BillAmount'].sum()
```

```python
a.isnull().sum()
```

```python
##6 fill missing value with mean
a.fillna({'BillAmount':a.BillAmount.mean()},inplace=True)
a
```

```python
##7 merge two datasets
b=pd.merge(a,df2,on='PatientID',how='inner')
```
